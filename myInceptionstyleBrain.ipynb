## Setup

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
```

## Prepare the data

```python
# Model / data parameters
num_classes = 10
input_shape = (28, 28, 1)

# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)
print("x_train shape:", x_train.shape)
print(x_train.shape[0], "train samples")
print(x_test.shape[0], "test samples")


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

## Build the model

```python
model1 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
  
        layers.Dense(64, activation='relu'),
        layers.Flatten(),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
```

```python
inputs = keras.Input(shape=input_shape)

layer1 = layers.Conv2D(32, kernel_size=(3, 3) )
layer2 = layers.MaxPooling2D(pool_size=(2, 2))
layer3  =layers.Conv2D(64, kernel_size=(3, 3) )
layer4 = layers.MaxPooling2D(pool_size=(2, 2))
layer5 = layers.Flatten()
layer6 = layers.Dense(num_classes, activation="softmax")


features2 = layer1 (inputs)
features3 = layer2 (features2)
features4 = layer3 (features3)
features5 = layer4 (features4)
features6 = layer5 (features5)
features7 = layer6 (features6)
 
model7 = keras.Model(inputs=inputs, outputs=features7)
```

```python
inputs = keras.Input(shape=input_shape)

layer1 = layers.Conv2D(32, kernel_size=(3, 3) )
layer2 = layers.MaxPooling2D(pool_size=(2, 2))
layer3  =layers.Conv2D(64, kernel_size=(3, 3) )
layer4 = layers.MaxPooling2D(pool_size=(2, 2))
layer5 = layers.Flatten()


layer11 = layers.Conv2D(32, kernel_size=(3, 3) )
layer12 = layers.MaxPooling2D(pool_size=(2, 2))
layer13  =layers.Conv2D(64, kernel_size=(3, 3) )
layer14 = layers.MaxPooling2D(pool_size=(2, 2))
layer15 = layers.Flatten()

features2 = layer1 (inputs)
features3 = layer2 (features2)
features4 = layer3 (features3)
features5 = layer4 (features4)
features6 = layer5 (features5)

features12 = layer11 (inputs)
features13 = layer12 (features12)
features14 = layer13 (features13)
features15 = layer14 (features14)
features16 = layer15 (features15)

featuresconctact = layers.Concatenate()([features6, features16])

layerfinal = layers.Dense(num_classes, activation="softmax")

outputfeatures = layerfinal (featuresconctact)

model8 = keras.Model(inputs=inputs, outputs=outputfeatures)
```

```python
model2 = keras.Sequential(
    [
        keras.Input(shape=input_shape),

        layers.Dense(64, activation='relu'),
 


        layers.Dense(32, activation='relu'),
             layers.BatchNormalization(),

        layers.Dense(32, activation='relu'),
 

        layers.Dense(32, activation='relu'),
        layers.Flatten(),
        layers.Dense(32, activation='relu'),
             layers.BatchNormalization(),

       
                  layers.Dropout(0.5),

        layers.Dense(32, activation='relu'),

     

        layers.Dense(num_classes, activation="softmax"),
    ]
)
```

```python
model3 = keras.Sequential(
    [
        keras.Input(shape=input_shape),

        layers.Dense(64, activation='relu'),
        layers.Flatten(),
        layers.Dense(32, activation='relu'),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
```

```python
model4 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3) ),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3) ),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
```

```python
model5 = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
```

## Train the model

```python
batch_size = 128
epochs = 2

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

log = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.3)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Retrieve a list of accuracy results on training and test data
# sets for each training epoch

history = log

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

# Retrieve a list of list results on training and test data
# sets for each training epoch
loss = history.history['loss']
val_loss = history.history['val_loss']

# Get number of epochs
epochs = range(len(acc))

# Plot training and validation accuracy per epoch
plt.plot(epochs, acc, 'C1', label='Training')
plt.plot(epochs, val_acc, 'C0', label='Validation')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()

```

```python
# Plot training and validation loss per epoch
plt.plot(epochs, loss,  'C1', label='Training loss')
plt.plot(epochs, val_loss , 'C0', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.figure()

```

## Evaluate the trained model

```python
score = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

```python
print("Test accuracy in percentage:", score[1] * 100)
```
